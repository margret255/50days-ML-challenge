```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
customer_data=pd.read_csv('Mall_Customers.csv')
```

```python
#chosing the anual income coumn and the spending score column
X=customer_data.iloc[:,[3,4]].values
```

```python
print(X)
```

```python
#choosing the correct no of clusters wcss values for different no of clusters
wcss=[]

for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)
```

```python
#elbow graph
sns.set()
if wcss: # Check if wcss is not empty
  plt.plot(range(1,11),wcss)
  plt.title("The elbow point graph")
  plt.xlabel("Number of clusters")
  plt.ylabel("WCSS")
  plt.show()
else:
  print("WCSS data is not available to plot the elbow graph.")
```

```python
kmeans=KMeans(n_clusters=5,init='k-means++',random_state=42)
y_kmeans=kmeans.fit_predict(X)
print(y_kmeans)
```

```python
#plotting the cluster and theri centroids
plt.figure(figsize=(8,8) )
plt.scatter(X[y_kmeans==0,0],X[y_kmeans==0,1],s=100,c='red',label='Cluster 1')
plt.scatter(X[y_kmeans==1,0],X[y_kmeans==1,1],s=100,c='blue',label='Cluster 2')
plt.scatter( X[y_kmeans==2,0],X[y_kmeans==2,1],s=100,c='green',label='Cluster 3')
plt.scatter(X[y_kmeans==3,0],X[y_kmeans==3,1],s=100,c='cyan',label='Cluster 4')
plt.scatter(X[y_kmeans==4,0],X[y_kmeans==4,1],s=100,c='magenta',label='Cluster 5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='yellow',label='Centroids')
plt.title("Customer groups")
plt.xlabel("Annual income")
plt.ylabel("Spending score")
plt.show()
```
